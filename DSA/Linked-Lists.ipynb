# 🔗 Linked Lists

## 📌 What is a Linked List?

A **Linked List** is a linear data structure made up of **nodes** that are connected together, forming a chain.

Unlike arrays, linked lists **do not store elements in contiguous memory**. Each node can exist anywhere in memory and is connected through references (pointers).

Example:

```text
Head
 │
 ▼
+------+------+
|  10  |  •───┼────►
+------+------+
             │
             ▼
        +------+------+
        |  20  |  •───┼────►
        +------+------+
                     │
                     ▼
                +------+------+
                |  30  | None |
                +------+------+
```

---

# 🧱 Node (Building Block)

A **Node** is the fundamental building block of a linked list.

Each node contains two parts:

- **Value (Data)** → The information being stored.
- **Next** → A reference (pointer) to the next node.

Python representation:

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next
```

---

## 📦 Node Components

### 1️⃣ Value (Data)

Stores the actual information.

Examples:

```python
10
```

```python
"Apple"
```

```python
3.14
```

---

### 2️⃣ Next Pointer

Stores the **reference to the next node**.

Example:

```
10 → 20 → 30
```

Here,

- Node 10 points to Node 20
- Node 20 points to Node 30

---

## 🏁 End of the List

The **last node** does not point to another node.

Instead, its **Next** pointer stores:

```python
None
```

Example:

```
10 → 20 → 30 → None
```

The `None` value tells Python:

> "This is the last node."

---

# 🧠 Memory Layout

Unlike arrays, linked list nodes are **not stored next to each other** in memory.

### Array

```
Memory

100
104
108
112
```

Everything is stored continuously.

---

### Linked List

```
Memory

100 → Node A

520 → Node B

812 → Node C
```

The nodes may exist in completely different memory locations.

Their **Next** pointers connect them together.

---

# 🔄 How Nodes Connect

Suppose we write:

```python
head = Node(10)
A = Node(20, head)
B = Node(30, A)
```

Memory:

```
B
 │
 ▼
30 ─────► 20 ─────► 10 ─────► None
```

Notice that the **second argument** tells each node **what it should point to**.

---

# 📚 Types of Linked Lists

There are two common types.

---

# 1️⃣ Singly Linked List

Each node stores **only one pointer**.

```
Head
 │
 ▼
10 → 20 → 30 → 40 → None
```

Each node knows only its **next** node.

### Structure

```python
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next
```

### Advantages

- Less memory
- Simple implementation
- Fast insertion at the beginning

### Disadvantages

- Cannot move backward
- Reverse traversal is difficult

---

# 2️⃣ Doubly Linked List

Each node stores **two pointers**.

- Previous
- Next

```
None ← 10 ⇄ 20 ⇄ 30 ⇄ 40 → None
```

Each node knows both:

- Previous node
- Next node

### Structure

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None
```

### Advantages

- Traverse forward
- Traverse backward
- Easier deletion

### Disadvantages

- Extra memory required
- Slightly more complex

---

# 🎯 Head

The **Head** is the reference to the **first node** in the linked list.

```
Head
 │
 ▼
10 → 20 → 30 → None
```

Without the Head, there is no way to access the list.

---

# 🏁 Tail

The **Tail** is the reference to the **last node**.

```
Head

↓

10 → 20 → 30 → 40

                 ↑

               Tail
```

Having a Tail pointer makes operations like appending new nodes much faster.

---

# ⚖️ Arrays vs Linked Lists

| Feature | Array | Linked List |
|----------|:-----:|:-----------:|
| Memory | Contiguous | Non-contiguous |
| Random Access | ✅ O(1) | ❌ O(n) |
| Insert at Beginning | ❌ O(n) | ✅ O(1) |
| Delete at Beginning | ❌ O(n) | ✅ O(1) |
| Insert in Middle | ❌ O(n) | ⚠️ O(n)\* |
| Extra Memory | None | Pointer(s) per node |
| Cache Friendly | ✅ Yes | ❌ No |

> **Note:** Insertion in the middle of a linked list is **O(1)** *only if you already have a reference to the insertion point*. Finding that position usually takes **O(n)**.

---

# ⏱️ Time Complexity

| Operation | Singly Linked List |
|-----------|:------------------:|
| Access by Index | **O(n)** |
| Search | **O(n)** |
| Insert at Beginning | **O(1)** |
| Insert at End (Tail available) | **O(1)** |
| Insert at End (No Tail) | **O(n)** |
| Delete at Beginning | **O(1)** |
| Delete at End | **O(n)** |

---

# 📝 Key Takeaways

- A **Linked List** is a chain of connected nodes.
- Each node stores:
  - **Value (Data)**
  - **Next Pointer**
- The last node points to **`None`**.
- Nodes are stored in **non-contiguous memory**.
- **Singly Linked Lists** have one pointer (`next`).
- **Doubly Linked Lists** have two pointers (`prev` and `next`).
- **Head** points to the first node.
- **Tail** points to the last node.
- Linked lists are excellent for **frequent insertions and deletions**, while arrays are better for **fast random access**.

````python
#Singly Node

class SinglyNode:

    def __init__(self,val,next=None):
        self.val = val
        self.next = next

    def __str__(self):
        return str(self.val)
````

````python
Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)

Head.next = A
A.next = B
B.next = C  

print(Head)
````

````python
#Traverse the list - O(n)

curr = Head

while curr:
    print(curr)
    curr = curr.next
````

````python
# Display Linked list - O(n)

def Display(head):
    curr = head
    elements = []

    while curr:
        elements.append(str(curr.val))
        curr = curr.next
    print(' --> '.join(elements))

Display(Head)
````

````python
# Doubly Linked Lists

class DoublyNode:
  def __init__(self, val, next=None, prev=None):
    self.val = val
    self.next = next
    self.prev = prev

  def __str__(self):
    return str(self.val)
````

````python
head = tail = DoublyNode(1)
print(tail)
````

````python
# Display - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' <-> '.join(elements))

display(head)
````

````python
# Insert at beginning - O(1)
def insert_at_beginning(head, tail, val):
  new_node = DoublyNode(val, next=head)
  head.prev = new_node
  return new_node, tail

head, tail = insert_at_beginning(head, tail, 3)
display(head)
````

````python
# Insert at end - O(1)
def insert_at_end(head, tail, val):
    new_node = DoublyNode(val, prev=tail)
    tail.next = new_node
    return head, new_node

head, tail = insert_at_end(head, tail, 7)
display(head)
````
