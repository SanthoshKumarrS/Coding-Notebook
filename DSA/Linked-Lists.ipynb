## Linked List

Definition: A linked list is a data structure consisting of nodes that point to other nodes, forming a chain

Node Components: A basic node (the building block) is an object containing:
Value: The data being stored (e.g., numbers or strings)

Next: A reference (or pointer) to the next node in the list

Ending the List: The last node points to a special marker, such as null or None in Python, to signify the end

Memory: Unlike arrays, linked lists are not stored in a contiguous sequence; nodes can be located at different, non-sequential memory addresses

Types of Linked Lists
Singly Linked List: Each node only has a connection to the next node

Doubly Linked List: Each node has connections to both the next and the previous node

Head and Tail:
 The Head is the reference to the front of the list

 The Tail is the reference to the back of the list (commonly used in doubly linked lists for easier backward traversal)



```python
#Singly Node

class SinglyNode:

    def __init__(self,val,next=None):
        self.val = val
        self.next = next

    def __str__(self):
        return str(self.val)
```

```python
Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)

Head.next = A
A.next = B
B.next = C  

print(Head)
```

```python
#Traverse the list - O(n)

curr = Head

while curr:
    print(curr)
    curr = curr.next
```

```python
# Display Linked list - O(n)

def Display(head):
    curr = head
    elements = []

    while curr:
        elements.append(str(curr.val))
        curr = curr.next
    print(' --> '.join(elements))

Display(Head)
```

```python
# Doubly Linked Lists

class DoublyNode:
  def __init__(self, val, next=None, prev=None):
    self.val = val
    self.next = next
    self.prev = prev

  def __str__(self):
    return str(self.val)
```

```python
head = tail = DoublyNode(1)
print(tail)
```

```python
# Display - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' <-> '.join(elements))

display(head)
```

```python
# Insert at beginning - O(1)
def insert_at_beginning(head, tail, val):
  new_node = DoublyNode(val, next=head)
  head.prev = new_node
  return new_node, tail

head, tail = insert_at_beginning(head, tail, 3)
display(head)
```

```python
# Insert at end - O(1)
def insert_at_end(head, tail, val):
    new_node = DoublyNode(val, prev=tail)
    tail.next = new_node
    return head, new_node

head, tail = insert_at_end(head, tail, 7)
display(head)
```
