# 🌳 Binary Tree Fundamentals

## 📌 What is a Binary Tree?

A **Binary Tree** is a hierarchical data structure made up of **nodes**, where each node can have **at most two children**:

- **Left Child**
- **Right Child**

Unlike linked lists, a node can branch into two different paths.

Example:

```text
          10
         /  \
        5    20
       / \   / \
      3   7 15 25
```

---

# 🧱 Binary Tree Node

Each node contains **three components**:

- **Value (Data)**
- **Left Child**
- **Right Child**

Python Representation:

```python
class TreeNode:

    def __init__(self, value):

        self.value = value
        self.left = None
        self.right = None
```

Example:

```
        10
       /  \
     None None
```

Initially, both children are:

```python
None
```

---

# 🌐 Binary Trees are Directed Graphs

A Binary Tree is a special type of **Directed Graph**.

Each connection (edge) points:

```
Parent

↓

Child
```

Example:

```
      10
     /  \
    5    20
```

You naturally move:

```
10

↓

5
```

or

```
10

↓

20
```

Traversal typically proceeds **from parent to child**, not upward.

---

# 📦 Array Representation

Binary Trees can also be stored inside an array.

For a node at index **i**:

```
Left Child

↓

2i
```

```
Right Child

↓

2i + 1
```

Example:

```
          A
         / \
        B   C
       / \
      D   E
```

Array Representation (1-based indexing):

| Index | Value |
|------:|:-----:|
| 1 | A |
| 2 | B |
| 3 | C |
| 4 | D |
| 5 | E |

Example:

Node:

```
B

Index = 2
```

Children:

```
Left = 4

Right = 5
```

The **Root** is the **top-most node** of the tree.

```

          ▼
         10
```

Every tree has exactly **one Root**.

Example:

```

       / \
      3   7
```

The subtree rooted at **5**:

```
      5
     / \
    3   7
```

---

## 🍃 Leaf Nodes

A **Leaf Node** has **no children**.

```
          10
         /  \
        5    20
       / \     \
      3   7     30
```

Leaf Nodes:

```
3

7

30
```

Both:

```
left = None

right = None
```

---

## 📏 Height (Maximum Depth)

The **Height** (or Maximum Depth) is the number of levels from the Root to the deepest Leaf Node.

Example:

```
          10
         /  \
        5    20
       /
      3
```

Levels:

```
Level 1

10

↓

Level 2

5

↓

Level 3

3
```

Height:

```

```

# 🌲 Types of Binary Trees

---

## 1️⃣ Perfect Binary Tree

Every level is **completely filled**.

```
          1
        /   \
       2     3
      / \   / \
     4  5  6   7
```

Properties:

- Every internal node has two children.
- Every leaf is at the same depth.

---

## 2️⃣ Complete Binary Tree

All levels are full **except possibly the last**, which is filled **from left to right**.

```
          1
        /   \
       2     3
      / \   /
     4  5  6
```

This is the structure used by **Binary Heaps**.

---

# 🔍 Tree Traversals

Traversal means **visiting every node** in a tree.

There are two major categories:

- Depth First Search (DFS)
- Breadth First Search (BFS)

---

# 🌳 Depth First Search (DFS)

DFS explores one branch **as deep as possible** before backtracking.

Usually implemented using:

- Recursion
- Stack

---

## 1️⃣ Preorder Traversal

Visit order:

```
Node

↓

Left

↓

Right
```

Example:

```
        A
       / \
      B   C
```

Traversal:

```
A

B

C
```

Python:

```python
def preorder(root):

    if root is None:
        return

    print(root.value)

    preorder(root.left)

    preorder(root.right)
```

## 2️⃣ Inorder Traversal

Visit order:

```
Left

↓

Node

↓

Right
```

Example:

```
        A
       / \
      B   C
```

Traversal:

```
B

A

C
```

---

## 3️⃣ Postorder Traversal

Visit order:

```
Left

↓

Right

↓

Node
```

Example:

```
        A
       / \
      B   C
```

Traversal:

```
B

C

A
```

# 🌊 Breadth First Search (BFS)

Also called:

```
Level Order Traversal
```

Nodes are visited **level by level**, from **top to bottom** and **left to right**.

Uses a **Queue**.

Example:

```
          10
         /  \
        5    20
       / \   / \
      3   7 15 25
```

Traversal:

```
10

5

20

3

7

15

25
```

# 🌟 Binary Search Tree (BST)

A **Binary Search Tree (BST)** is a special type of Binary Tree.

For **every node**:

```
Left Subtree

< Node
```

```
Right Subtree

> Node
```

Example:

```
           15
         /    \
       10      20
      /  \    /  \
     8   12 18   25
```

Notice:

```
Left values

< 15
```

```
Right values

> 15
```

This rule applies **recursively** to every node.

---

# 🔍 Searching in a BST

Suppose we search for:

```
18
```

```
           15
             \
             20
            /
          18
```

Search process:

```
18 > 15

↓

Go Right

↓

18 < 20

↓

Go Left

↓

Found
```

Each comparison eliminates roughly **half of the remaining tree**, making the search efficient when the tree is balanced.

---

# ⭐ Inorder Traversal Advantage

One special property of a BST:

An **Inorder Traversal** always visits the nodes in **sorted order**.

Example:

```
           15
         /    \
       10      20
      /  \    /  \
     8   12 18   25
```

Inorder:

```
8

10

12

15

18

20

25
```

Perfectly sorted.

---

# ⏱️ Time Complexity

## General Binary Tree Search

There is **no ordering**.

Worst case:

Every node must be visited.

```
Time Complexity = O(n)
```

---

## Balanced BST Search

Each comparison removes approximately **half of the remaining tree**.

```
Time Complexity = O(log n)
```

---

## Worst-Case BST

If the BST becomes skewed:

```
10
  \
   20
     \
      30
        \
         40
```

It behaves like a **Linked List**.

Searching becomes:

```
O(n)
```

---

# 💾 Space Complexity

---

## DFS

Uses the **Call Stack** (or an explicit stack).

Worst Case:

```
O(n)
```

---

## BFS

Uses a **Queue**.

Worst Case:

```
O(n)
```

---

# 📊 Complexity Summary

| Operation | Binary Tree | Balanced BST | Skewed BST |
|-----------|:-----------:|:------------:|:----------:|
| Search | **O(n)** | **O(log n)** | **O(n)** |
| DFS Space | **O(n)** | **O(n)** | **O(n)** |
| BFS Space | **O(n)** | **O(n)** | **O(n)** |

---

# 🌲 DFS vs BFS

| Feature | DFS | BFS |
|----------|:---:|:---:|
| Traversal Style | Deep First | Level by Level |
| Data Structure | Stack / Recursion | Queue |
| Space Complexity | O(n) | O(n) |
| Common Uses | Trees, Backtracking, DFS | Shortest Path, Level Order |

---

# 📝 Key Takeaways

- A **Binary Tree** is made of nodes, each having at most **two children**.
- Each node contains:
  - **Value**
  - **Left Child**
  - **Right Child**
- The **Root** is the top-most node, and any node can serve as the root of a **Subtree**.
- **Leaf Nodes** have no children.
- The **Height (Maximum Depth)** is the number of levels from the root to the deepest leaf.
- **Perfect Binary Trees** have every level completely filled.
- **Complete Binary Trees** have all levels full except possibly the last, which is filled from left to right.
- **Depth First Search (DFS)** includes:
  - Preorder → Node → Left → Right
  - Inorder → Left → Node → Right
  - Postorder → Left → Right → Node
- **Breadth First Search (BFS)** visits nodes level by level using a **Queue**.
- A **Binary Search Tree (BST)** maintains:
  - Left Subtree < Node
  - Right Subtree > Node
- An **Inorder Traversal** of a BST always produces values in **sorted order**.
- Search complexity:
  - General Binary Tree → **O(n)**
  - Balanced BST → **O(log n)**
  - Skewed BST → **O(n)**
- DFS and BFS both require **O(n)** space in the worst case.

````python
# Binary Trees
class TreeNode:
  def __init__(self, val, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

  def __str__(self):
    return str(self.val)
````

````python
#        1
#     2    3
#   4  5  10
````

````python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
````

````python
# Recursive Pre Order Traversal (DFS) Time: O(n), Space: O(n)

def pre_order_Traversal(node):
    if not node:
        return 

    print(node)
    pre_order_Traversal(node.left)
    pre_order_Traversal(node.right)

pre_order_Traversal(A)
````

````python
# Recursive In Order Traversal (DFS) Time: O(n), Space: O(n)

def in_order_Traversal(node):
    if not node:
        return 

    
    in_order_Traversal(node.left)
    print(node)
    in_order_Traversal(node.right)

in_order_Traversal(A)
````

````python
# Iterative Pre Order Traversal (DFS) Time: O(n), Space: O(n)

def pre_order_Iterative(node):
    stk = [node]

    while stk:
        node = stk.pop()
        print(node)
        if node.right: stk.append(node.right)
        if node.left: stk.append(node.left)

pre_order_Iterative(A)
````

````python
# Level Order Traversal (BFS) Time: O(n), Space: O(n)

from collections import deque

def level_order_Traversal(node):
    q = deque()
    q.append(node)

    while q:
        node = q.popleft()
        print(node)
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)

level_order_Traversal(A)
````

````python
# Check if Value Exists (DFS) Time: O(n), Space: O(n)
def search(node, target):
  if not node:
    return False

  if node.val == target:
    return True

  return search(node.left, target) or search(node.right, target)

search(A, 11)
````
