# 🏔️ Heap & Priority Queue Fundamentals

## 📌 What is a Heap?

A **Heap** is a specialized **Binary Tree** used to efficiently manage elements based on **priority** rather than insertion order.

It is commonly implemented as a **Priority Queue**, where the highest-priority element is always processed first.

Unlike a Binary Search Tree (BST), a Heap does **not** maintain sorted order—it only guarantees that the highest (or lowest) priority element is at the root.

---

# 🎯 What is a Priority Queue?

A **Priority Queue** is a data structure where elements are removed based on their **priority**, not on when they were inserted.

Example:

Suppose smaller numbers have higher priority.

```
Insert:

10

5

20

1
```

Removal Order:

```
1

5

10

20
```

Notice:

Although **1** was inserted last, it is removed first because it has the highest priority.

---

# 🌳 Heap Structure

A Heap is a **Complete Binary Tree**.

Example (Min Heap):

```text
          2
        /   \
       5     8
      / \   / \
    10 15 20 25
```

Every level is completely filled except possibly the last, which is filled **from left to right**.

---

# 📦 Array Representation

Unlike ordinary binary trees, Heaps are efficiently stored in an **array**.

Example:

```text
          2
        /   \
       5     8
      / \   / \
    10 15 20 25
```

Array:

| Index | Value |
|------:|:-----:|
| 0 | 2 |
| 1 | 5 |
| 2 | 8 |
| 3 | 10 |
| 4 | 15 |
| 5 | 20 |
| 6 | 25 |

No pointers are needed.

---

# 📐 Heap Index Formula

For any node at index **i**:

### Left Child

```python
left = 2 * i + 1
```

### Right Child

```python
right = 2 * i + 2
```

### Parent

```python
parent = (i - 1) // 2
```

Example:

```
Index = 1

Value = 5
```

Children:

```
Left = 3

Right = 4
```

Values:

```
10

15
```

# 🎯 Priority Logic

In a Priority Queue:

The element with the **highest priority** is always removed first.

Example:

```
Emergency Room
```

```
Priority

```

Treatment Order:

```
1

```

Not based on arrival time.

# 📉 Min Heap

A **Min Heap** satisfies:

```
Parent ≤ Children
```

Example:

```text
          2
        /   \
       5     8
      / \   / \
    10 15 20 25
```

Notice:

```
2 < 5

2 < 8

5 < 10

5 < 15
```

The **minimum element is always the root**.

---

# 📈 Max Heap

A **Max Heap** satisfies:

```
Parent ≥ Children
```

Example:

```text
          25
        /    \
      20      18
     /  \    /  \
    10  15  8   5
```

The **maximum element is always the root**.

---

# 🔄 Sifting (Heap Adjustment)

Whenever an element is inserted or removed, the heap property may be violated.

To restore it, Heaps use:

- **Sift Up**
- **Sift Down**

---

## ⬆️ Sift Up

Used after **insertion**.

Example:

Insert:

```
1
```

Before:

```text
        2
      /   \
     5     8
```

Insert at the end:

```text
        2
      /   \
     5     8
    /
   1
```

Violation:

```
1 < 5
```

Swap:

```text
        2
      /   \
     1     8
    /
   5
```

Again:

```
1 < 2
```

Swap:

```text
        1
      /   \
     2     8
    /
   5
```

Heap restored.

---

## ⬇️ Sift Down

Used after **deletion**.

Suppose we remove:

```
1
```

Replace with last element:

```text
        25
      /   \
     2     8
```

Violation:

```
25 > 2
```

Swap:

```text
        2
      /   \
     25    8
```

Continue until the heap property is restored.

# ⚙️ Heap Operations

---

# 1️⃣ Heapify

Transforms an **unsorted array** into a valid Heap.

Example:

Before:

```
[8, 3, 10, 1, 5]
```

After Heapify:

```
        1
      /   \
     3     10
    / \
   8   5
```

### Time Complexity

```
O(n)
```

---

# 2️⃣ Heap Push

Insert a new element.

Process:

1. Add at the end.
2. Sift Up.

Example:

```
Insert 4
```

↓

```
Append

↓

Sift Up
```

### Time Complexity

```
O(log n)
```

---

# 3️⃣ Heap Pop

Remove the root.

Process:

1. Remove root.
2. Move last element to root.
3. Sift Down.

Example:

```
Remove Minimum
```

↓

```
Replace Root
```

↓

```
Sift Down
```

### Time Complexity

```
O(log n)
```

---

# 4️⃣ Heap Peek

View the highest-priority element without removing it.

Since the root is stored at index **0**:

```python
heap[0]
```

### Time Complexity

```
O(1)
```

---

# 📝 Python Heap (`heapq`)

Python provides a built-in **Min Heap** implementation.

```python
import heapq

heap = []
```

Insert:

```python
heapq.heappush(heap, 10)
heapq.heappush(heap, 5)
heapq.heappush(heap, 20)
```

Peek:

```python
heap[0]
```

Remove Minimum:

```python
heapq.heappop(heap)
```

# 🔃 Heap Sort

Heap Sort repeatedly removes the root of the heap.

Example:

```
Heap

↓

1

↓

2

↓

5

↓

8

↓

10
```

Each removed element is added to the sorted array.

Result:

```
[1, 2, 5, 8, 10]
```

---

# ⏱️ Heap Sort Complexity

Building Heap:

```
O(n)
```

Removing every element:

```
n × O(log n)
```

Overall:

```
O(n log n)
```

---

# 💾 Space Complexity

Simple Heap Sort:

```
O(n)
```

(using an additional array)

---

In-place Heap Sort:

```
O(1)
```

No additional array is required.

# 📊 Min Heap vs Max Heap

| Feature | Min Heap | Max Heap |
|----------|:--------:|:--------:|
| Root | Smallest Element | Largest Element |
| Parent Rule | Parent ≤ Children | Parent ≥ Children |
| Highest Priority | Smallest Value | Largest Value |
| Python `heapq` | ✅ Yes | ❌ (simulate using negative values) |

# 📝 Key Takeaways

- A **Heap** is a specialized **Complete Binary Tree** used to implement **Priority Queues**.
- Heaps are efficiently stored using an **array**, eliminating the need for pointers.
- Array relationships:
  - Left Child → `2 * i + 1`
  - Right Child → `2 * i + 2`
  - Parent → `(i - 1) // 2`
- **Min Heap:** Parent ≤ Children, so the smallest element is always at the root.
- **Max Heap:** Parent ≥ Children, so the largest element is always at the root.
- **Sift Up** restores the heap after insertion.
- **Sift Down** restores the heap after deletion.
- Common operations:
  - Heapify → **O(n)**
  - Peek → **O(1)**
  - Push → **O(log n)**
  - Pop → **O(log n)**
- **Heap Sort** repeatedly removes the root to produce a sorted sequence with **O(n log n)** time complexity.
- Python's **`heapq`** module provides an efficient **Min Heap** implementation.

````python
# Build Min Heap (Heapify)
# Time: O(n), Space: O(1)

A = [-4,-3,1,0,2,5,10,8,12,9]

import heapq
heapq.heapify(A)

A
````

````python
# Heap Push (Insert element)
# Time: O(log n)

heapq.heappush(A, 4)
A
````

````python
# Heap Pop (Extract min)
# Time: O(log n)

minn = heapq.heappop(A)

print(A, minn)
````

````python
# Heap Sort
# Time: O(n log n), Space: O(n)
# NOTE: O(1) Space is possible via swapping, but this is complex

def heap_sort(arr):
    heapq.heapify(arr)
    n = len(arr)
    new_list = [0] * n

    for i in range(n):
        minn = heapq.heappop(arr)
        new_list[i] = minn

    return new_list


heap_sort([1,4,2,-9,-6,0,8,10,52])
````

````python
# Heap Push Pop: Time: O(log n)

heapq.heappushpop(A, 99)

A
````

````python
# Peak at Min: Time O(1)
A[0]
````

````python
# Build heap from scratch - Time: O(n log n)

c = [-5,-9,0,6,9,8,1,3]

heap = []

for x in c:
    heapq.heappush(heap,x)
    print(f'Heap = {heap}, length = {len(heap)}')
````
