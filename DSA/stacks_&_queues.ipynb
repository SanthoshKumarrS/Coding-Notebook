# 📚 Stack (LIFO)

## 📌 What is a Stack?

A **Stack** is a linear data structure in which elements are **added and removed from the same end**, called the **Top**.

It follows the **LIFO (Last In, First Out)** principle.

Think of a stack of plates:

- 🍽️ The last plate placed on top is the first one removed.

---

# 🔄 LIFO (Last In, First Out)

The **last element inserted** is the **first element removed**.

Example:

```
Push 10

10
```

```
Push 20

20
10
```

```
Push 30

30
20
10
```

Now remove one element:

```
Pop

20
10
```

The last inserted value (**30**) is removed first.

---

# 🎯 Stack Visualization

Stacks are often visualized vertically.

```
Top
 │
 ▼
+------+
|  40  |
+------+
|  30  |
+------+
|  20  |
+------+
|  10  |
+------+
```

Only the **Top** element is accessible.

---

# ⚙️ Stack Operations

---

## 1️⃣ Push (Append)

Adds a new element to the **Top** of the stack.

Python:

```python
stack.append(50)
```

Example:

Before

```
Top

30
20
10
```

After

```
Top

50
30
20
10
```

### Time Complexity

**O(1)** (Average)

---

## 2️⃣ Pop

Removes the **Top** element.

Python:

```python
stack.pop()
```

Example:

Before

```
Top

40
30
20
```

After

```
Top

30
20
```

No elements need to be shifted.

### Time Complexity

**O(1)**

---

## 3️⃣ Peek

Returns the top element **without removing it**.

Python:

```python
stack[-1]
```

Example:

```
Top

40
30
20
```

Peek returns:

```
40
```

The stack remains unchanged.

### Time Complexity

**O(1)**

---

## 4️⃣ IsEmpty

Checks whether the stack contains any elements.

Python:

```python
if stack:
    print("Stack is not empty")
```

or

```python
len(stack) == 0
```

### Time Complexity

**O(1)**

---

# ⏱️ Stack Time Complexity

| Operation | Time Complexity |
|-----------|:---------------:|
| Push | **O(1)** (Average) |
| Pop | **O(1)** |
| Peek | **O(1)** |
| IsEmpty | **O(1)** |

Stacks are commonly used for:

- Undo / Redo Operations
- Browser History
- Expression Evaluation
- Parentheses Matching
- Depth-First Search (DFS)
- Backtracking Algorithms

# 📚 Queue (FIFO)

## 📌 What is a Queue?

A **Queue** is a linear data structure in which elements are:

- Added at the **Rear (Back)**
- Removed from the **Front**

It follows the **FIFO (First In, First Out)** principle.

Think of people waiting in a queue.

The first person to enter the line is the first one served.

---

# 🔄 FIFO (First In, First Out)

The **first inserted element** is the **first removed**.

Example:

```
Enqueue 10

10
```

```
Enqueue 20

10 20
```

```
Enqueue 30

10 20 30
```

Now remove one element:

```
Dequeue

20 30
```

The oldest element (**10**) leaves first.

---

# 🎯 Queue Visualization

```
Front                         Rear

10 → 20 → 30 → 40
```

New elements enter from the **Rear**.

Elements leave from the **Front**.

---

# ⚙️ Queue Operations

---

## 1️⃣ Enqueue

Adds an element to the **Rear**.

Python (`deque`):

```python
queue.append(50)
```

Example:

Before

```
Front

10 20 30
```

After

```
Front

10 20 30 50

           Rear
```

### Time Complexity

**O(1)**

---

## 2️⃣ Dequeue

Removes the element from the **Front**.

Python (`deque`):

```python
queue.popleft()
```

Example:

Before

```
Front

10 20 30
```

After

```
Front

20 30
```

### Time Complexity

**O(1)**

---

## ⚠️ Why Not Use a Python List?

Suppose we have:

```python
numbers = [10, 20, 30, 40]
```

Removing the first element:

```python
numbers.pop(0)
```

Python must shift every remaining element.

```
Before

10 20 30 40
```

↓

```
20 30 40
```

Everything moves one position left.

Time Complexity:

**O(n)**

Therefore, **Python lists are inefficient for queues**.

---

# ✅ Python's deque

Python provides an optimized queue implementation:

```python
from collections import deque

queue = deque()
```

Operations:

```python
queue.append(10)      # Enqueue

queue.popleft()       # Dequeue
```

Both operations are:

**O(1)**

---

# 🔄 Why deque is Fast

A `deque` (Double-Ended Queue) is specifically designed for efficient insertion and deletion at **both ends**.

Unlike Python lists, it does **not** shift elements when removing from the front.

---

# ⏱️ Queue Time Complexity

| Operation | Python List | deque |
|-----------|:-----------:|:-----:|
| Enqueue | **O(1)** | **O(1)** |
| Dequeue | **O(n)** | **O(1)** |
| Peek Front | **O(1)** | **O(1)** |
| IsEmpty | **O(1)** | **O(1)** |

---

# 📊 Stack vs Queue

| Feature | Stack | Queue |
|---------|:-----:|:-----:|
| Principle | LIFO | FIFO |
| Insert | Top | Rear |
| Remove | Top | Front |
| Python Implementation | `list` | `collections.deque` |
| Push/Enqueue | `append()` | `append()` |
| Pop/Dequeue | `pop()` | `popleft()` |

# 📝 Key Takeaways

## Stack

- Follows **LIFO (Last In, First Out)**.
- Elements are added and removed from the **Top**.
- Main operations:
  - Push (`append()`)
  - Pop (`pop()`)
  - Peek (`stack[-1]`)
  - IsEmpty
- Most operations run in **O(1)** time.

---

## Queue

- Follows **FIFO (First In, First Out)**.
- Elements are added at the **Rear** and removed from the **Front**.
- Main operations:
  - Enqueue (`append()`)
  - Dequeue (`popleft()`)
- Use **`collections.deque`** in Python for efficient **O(1)** queue operations.
- Avoid using a regular Python list as a queue because removing from the front (`pop(0)`) takes **O(n)** time.

````python
#Stacks

stk = []

stk.append('a')
stk.append('b')
stk.append('c')

print(stk)
````

````python
#pop from the stack

x = stk.pop()

print(x)
print(stk)
````

````python
# Ask what's on the top of stack - O(1)
print(stk[-1])
````

````python
# Queues 
from collections import deque

q = deque()

#enqueue
q.append('a')
q.append('b')
q.append('c')

q
````

````python
#deque (popleft)

q.popleft()
````

````python
# Peek from left side - O(1)
q[0]
````

````python
# Peek from right side - O(1)
q[-1]
````
