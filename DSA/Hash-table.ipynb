# 🗂️ Hash Table Fundamentals

## 📌 Definition

A **Hash Table** is a data structure that stores data by mapping a **key** (such as a string or integer) to a specific index in an array called **buckets**.

> **Purpose:** Enable extremely fast insertion, lookup, and deletion of data.

---

## ⚙️ Hash Function

A **Hash Function** converts an input (key) into an integer index.

Example:

```python
index = hash(key) % number_of_buckets
```

The **modulo operator (`%`)** ensures the generated index stays within the valid range of available buckets.

**Example:**

```python
sum = 27
buckets = 5

index = 27 % 5 = 2
```

The item is stored in **Bucket 2**.

---

## 🎯 Goal of a Hash Table

Hash Tables are primarily used to implement:

- **Hash Set** → Stores **unique** elements.
- **Hash Map (Dictionary)** → Stores **key-value pairs**.

These structures provide **very fast data retrieval**.

---

# ⚠️ Handling Collisions

## What is a Collision?

A **collision** occurs when **two different keys produce the same hash index**.

Example:

```
Apple  ──► Bucket 3
Mango  ──► Bucket 3
```

Both keys map to the same bucket.

---

## 1️⃣ Separate Chaining

In **Separate Chaining**, each bucket contains a **linked list**.

When a collision occurs, the new element is simply appended to the linked list.

```
Bucket 0

Head
 │
 ▼
+--------+      +--------+      +--------+
| Apple  | ───► | Mango  | ───► | Orange | ───► None
+--------+      +--------+      +--------+
```

### ✅ Advantages

- Simple to implement
- Easy insertion and deletion
- Table never becomes completely full

### ❌ Disadvantages

- Extra memory required for linked lists
- Many collisions can reduce performance

---

## 2️⃣ Linear Probing

Instead of storing a linked list, the algorithm searches for the **next available empty bucket**.

Example:

```
Original Index = 3

Bucket 3 → Occupied ❌
Bucket 4 → Occupied ❌
Bucket 5 → Empty ✅

Store item in Bucket 5
```

This process continues until an empty bucket is found.

### ✅ Advantages

- No extra memory needed
- Better cache performance

### ❌ Disadvantages

- Clustering can occur
- Performance decreases as the table fills up

---

## 🗑️ Deletion in Linear Probing

When deleting an element, **do not simply make the bucket empty**.

Instead, mark it with a special value (often called a **tombstone**).

Example:

```
Before

[10] [25] [37] [Empty]

Delete 25

[10] [-1] [37] [Empty]
```

Why?

If the bucket were marked as **Empty**, future searches might stop early and incorrectly conclude that later elements don't exist.

---

# ⏱️ Time Complexity (Big-O)

| Operation | Average Case | Worst Case |
|-----------|:------------:|:----------:|
| Search | **O(1)** | **O(n)** |
| Insert | **O(1)** | **O(n)** |
| Delete | **O(1)** | **O(n)** |

---

## 🚀 Average Case — O(1)

A good hash function distributes keys evenly across buckets.

The algorithm directly computes the bucket index and accesses the element in **constant time**.

```
Key
 ↓
Hash Function
 ↓
Bucket Index
 ↓
Retrieve Item
```

---

## ⚠️ Worst Case — O(n)

In the worst case, **every key hashes to the same bucket**.

```
Bucket 2

Head
 │
 ▼
A → B → C → D → E → F → G
```

Searching now requires traversing the entire list, resulting in **O(n)** time.

---

# ⚖️ Space-Time Tradeoff

Hash Tables trade **additional memory** for **faster operations**.

### Compared with an Array

| Data Structure | Lookup |
|---------------|:------:|
| Array (Linear Search) | **O(n)** |
| Hash Table (Average) | **O(1)** |

To maintain high performance:

- ✔️ Use a well-designed hash function.
- ✔️ Keep enough buckets to minimize collisions.
- ✔️ Resize (rehash) the table when it becomes too full.

---

# 📝 Key Takeaways

- A **Hash Table** maps keys to bucket indices using a **hash function**.
- The **modulo operator (`%`)** keeps indices within the bucket range.
- **Collisions** happen when multiple keys map to the same bucket.
- Common collision handling techniques:
  - Separate Chaining
  - Linear Probing
- Hash Tables power:
  - Hash Sets
  - Hash Maps (Dictionaries)
- Average-case complexity for search, insertion, and deletion is **O(1)**.
- Worst-case complexity is **O(n)** when collisions are excessive.
- Performance depends on:
  - A good hash function
  - Low collision rate
  - Proper table sizing
