# 🔁 Recursion Fundamentals

## 📌 What is Recursion?

**Recursion** is a programming technique where a **function calls itself** to solve a problem by breaking it into **smaller instances of the same problem**.

Instead of using loops, recursion repeatedly calls itself until it reaches a stopping condition called the **Base Case**.

---

# 🎯 Basic Structure of Recursion

Every recursive function has **two essential parts**:

1. **Base Case**
2. **Recursive Step**

```python
def recursive_function(input):

    # Base Case
    if stopping_condition:
        return

    # Recursive Step
    return recursive_function(smaller_input)
```

---

# 🛑 Base Case

The **Base Case** is the condition that stops recursion.

Without a base case, the function would continue calling itself forever, leading to a **RecursionError (Stack Overflow)**.

Example:

```python
def countdown(n):
    if n == 0:
        print("Done!")
        return

    print(n)
    countdown(n - 1)
```

Execution:

```
countdown(3)

↓

3

↓

2

↓

1

↓

Done!
```

The recursion stops when:

```python
n == 0
```

---

# 🔄 Recursive Step

The **Recursive Step** is where the function calls itself with a **smaller or simpler input**, moving closer to the base case.

Example:

```python
countdown(n - 1)
```

Each call reduces:

```
3

↓

2

↓

1

↓

0
```

Eventually, the base case is reached.

---

# 🌳 Recursive Data Structures

Many data structures are naturally recursive.

Examples:

### Linked List

```
Node

↓

Node

↓

Node

↓

None
```

Each node points to **another node**.

---

### Tree

```
        A
      /   \
     B     C
    / \   / \
   D   E F   G
```

Each node can have child nodes, which are themselves trees.

---

### Graph

Graphs can also be traversed recursively using:

- Depth-First Search (DFS)
- Backtracking

---

# 📚 Function Call Stack

Every recursive function call is stored in a **Call Stack**.

Each function call creates a new **Stack Frame**.

---

## What Does a Stack Frame Store?

Each frame stores:

- Function parameters
- Local variables
- Return address

Example:

```python
factorial(3)
```

Call Stack:

```
factorial(3)

↓

factorial(2)

↓

factorial(1)

↓

factorial(0)
```

Each call waits until the next call finishes.

---

# 🔄 Stack Unwinding

Once the **Base Case** is reached, recursion starts returning values back up the stack.

Example:

```python
factorial(3)
```

Call Stack:

```
factorial(3)

↓

factorial(2)

↓

factorial(1)

↓

factorial(0)
```

Base Case:

```
return 1
```

Now the stack unwinds:

```
factorial(1)

↓

1 × 1 = 1

↓

factorial(2)

↓

2 × 1 = 2

↓

factorial(3)

↓

3 × 2 = 6
```

Final Answer:

```
6
```

---

# 🔢 Fibonacci Sequence

The Fibonacci sequence is defined as:

```
F(0) = 0

F(1) = 1
```

For every other number:

```
F(n) = F(n-1) + F(n-2)
```

Python:

```python
def fib(n):

    if n == 0:
        return 0

    if n == 1:
        return 1

    return fib(n-1) + fib(n-2)
```

---

# 🌳 Recursive Tree

Calling:

```python
fib(5)
```

creates this recursion tree:

```
                 fib(5)
               /        \
          fib(4)       fib(3)
          /    \       /    \
     fib(3) fib(2) fib(2) fib(1)
      / \      / \     / \
...
```

Each node generates **two more recursive calls**, creating a tree-like structure.

---

# ⚠️ Why is Recursive Fibonacci Slow?

Many values are calculated repeatedly.

Example:

```
fib(5)

↓

fib(4)

↓

fib(3)

↓

fib(2)
```

Later:

```
fib(5)

↓

fib(3)

↓

fib(2)
```

Notice:

```
fib(2)
```

is computed **multiple times**.

This repeated work makes the naive recursive Fibonacci algorithm inefficient.

# 🔗 Example: Reversing a Linked List (Printing in Reverse)

Suppose we have:

```
10 → 20 → 30 → None
```

Recursive idea:

```
Visit 10

↓

Visit 20

↓

Visit 30

↓

None
```

Nothing is printed while moving forward.

Once the end is reached:

```
Print 30

↓

Print 20

↓

Print 10
```

Output:

```
30

20

10
```

The linked list itself is **not modified**.

The reverse order appears because of **stack unwinding**.

# ⏱️ Time Complexity

## Fibonacci

Each function call creates two more calls.

```
fib(5)

↓

fib(4)

↓

fib(3)

↓

fib(2)
```

The number of calls grows exponentially.

### Time Complexity

```
O(2ⁿ)
```

Very slow for large values of **n**.

---

## Linked List Traversal

Each node is visited exactly once.

```
10

↓

20

↓

30

↓

40
```

### Time Complexity

```
O(n)
```

---

# 💾 Space Complexity

Recursion uses memory for the **Call Stack**.

General Rule:

```
Space Complexity = O(Depth of Recursion)
```

---

## Fibonacci

Maximum recursion depth:

```
n
```

Therefore:

```
Space Complexity = O(n)
```

---

## Linked List Reverse

Each recursive call processes one node.

```
10

↓

20

↓

30

↓

40
```

Before returning, all function calls remain on the stack.

### Space Complexity

```
O(n)
```

# 📝 Key Takeaways

- **Recursion** is when a function calls itself.
- Every recursive function must have:
  - ✅ A **Base Case**
  - ✅ A **Recursive Step**
- The **Base Case** prevents infinite recursion.
- Each recursive call creates a **Stack Frame** in the **Call Stack**.
- Once the Base Case is reached, the stack begins **unwinding**, returning values back to previous calls.
- **Linked Lists, Trees, and Graphs** are naturally recursive data structures.
- The naive recursive Fibonacci algorithm is inefficient because it performs **duplicate calculations**, resulting in **O(2ⁿ)** time complexity.
- Recursive linked list traversal visits each node once, giving **O(n)** time complexity.
- The space complexity of recursion is generally **O(recursion depth)** due to the function call stack.

````python
#Fibonacci Sequence
#F(0) = 0
#F(1) = 1
# n >1: F(n) = F(n-1) + F(n-2)
#Time: O(2^n), Space: O(n)

def F(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return F(n-1) +F(n-2)

F(15)
````

````python
#Linked Lists

class SinglyNode:

  def __init__(self, val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)

Head = SinglyNode(1)
A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(7)

Head.next = A
A.next = B
B.next = C

print(Head)
````

````python
# Time: O(n), Space: O(n)
def reverse(node):
  if not node:
    return

  reverse(node.next)
  print(node)


reverse(Head)
````
