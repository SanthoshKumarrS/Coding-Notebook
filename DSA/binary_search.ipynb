# 🔍 Binary Search Fundamentals

## 📌 What is Binary Search?

**Binary Search** is an efficient searching algorithm used to find a target element in a **sorted array**.

Instead of checking every element one by one (Linear Search), Binary Search repeatedly **divides the search space in half**, making it much faster for large datasets.

---

# 🎯 Why Use Binary Search?

### Linear Search

```
Target = 45

10 20 30 40 45 50 60

↓

Check 10

↓

Check 20

↓

Check 30

↓

Check 40

↓

Check 45 ✅
```

Worst Case:

```
O(n)
```

---

### Binary Search

```
Target = 45

10 20 30 40 45 50 60
```

Check the middle first.

```
40
```

Since:

```
45 > 40
```

Discard the entire left half.

Continue searching only in the right half.

```
45 50 60
```

Very few comparisons are required.

---

# ✅ Requirement

**Binary Search only works on sorted data.**

✔️ Correct:

```
10 20 30 40 50 60
```

❌ Incorrect:

```
20 50 10 30 60
```

If the array is not sorted, Binary Search cannot determine which half to eliminate.

---

# ⚙️ Initial Setup

Binary Search uses two pointers.

```
L → Left Pointer

R → Right Pointer
```

Initially:

```python
L = 0
R = len(arr) - 1
```

Example:

```
Index

0 1 2 3 4 5 6

Array

10 20 30 40 50 60 70
```

Pointers:

```
L                         R

↓

10 20 30 40 50 60 70
```

---

# 📍 Finding the Middle Index

The middle index is calculated using:

```python
M = (L + R) // 2
```

The `//` operator performs **integer division**, always returning an integer.

Example:

```python
L = 0
R = 6

M = (0 + 6) // 2

M = 3
```

Middle element:

```
40
```

---

# 🔄 Binary Search Algorithm

At each iteration:

1. Find the middle element.
2. Compare it with the target.
3. Eliminate half of the search space.

---

## Case 1: Target Found

```python
arr[M] == target
```

Example:

```
Target = 40

10 20 30 40 50 60
```

Middle:

```
40
```

Found!

Return the index (or `True`).

---

## Case 2: Target is Smaller

```python
target < arr[M]
```

Example:

```
Target = 20

10 20 30 40 50 60
```

Middle:

```
40
```

Since:

```
20 < 40
```

Discard the entire right half.

Update:

```python
R = M - 1
```

New search range:

```
10 20 30
```

---

## Case 3: Target is Larger

```python
target > arr[M]
```

Example:

```
Target = 60

10 20 30 40 50 60
```

Middle:

```
40
```

Since:

```
60 > 40
```

Discard the left half.

Update:

```python
L = M + 1
```

New search range:

```
50 60
```

---

# 🛑 Termination Condition

Continue searching while:

```python
L <= R
```

If the pointers cross:

```python
L > R
```

The target does **not** exist.

Example:

```
L

↓

10 20 30

          ↑

          R
```

Eventually:

```
L > R
```

Stop searching.

Return:

```python
False
```

or

```python
-1
```

---

# ⚠️ Integer Overflow

In some programming languages:

```python
M = (L + R) // 2
```

may cause an **integer overflow** if:

```
L

=

2,000,000,000

R

=

2,000,000,000
```

Adding them exceeds the maximum integer size.

---

# ✅ Safer Formula

To avoid overflow:

```python
M = L + ((R - L) // 2)
```

Why it works:

Instead of adding two large numbers:

```
L + R
```

We calculate:

```
Distance between pointers

↓

Half of that distance

↓

Add it back to L
```

It produces the same midpoint while avoiding overflow.

> **Note:** Python integers automatically grow as needed, so integer overflow is generally **not an issue in Python**. However, this safer formula is commonly used in languages like C++ and Java.

# 📝 Binary Search Algorithm (Python)

```python
def binary_search(arr, target):

    L = 0
    R = len(arr) - 1

    while L <= R:

        M = (L + R) // 2

        if arr[M] == target:
            return M

        elif target < arr[M]:
            R = M - 1

        else:
            L = M + 1

    return -1
```

---

# 📈 Binary Search vs Linear Search

| Feature | Linear Search | Binary Search |
|----------|:-------------:|:-------------:|
| Requirement | Unsorted or Sorted | **Sorted Only** |
| Time Complexity | **O(n)** | **O(log n)** |
| Space Complexity | **O(1)** | **O(1)** |
| Best For | Small / Unsorted Data | Large Sorted Data |

---

# 📝 Key Takeaways

- **Binary Search** works only on **sorted arrays**.
- It repeatedly **cuts the search space in half**, making it much faster than Linear Search.
- Use two pointers:
  - **L (Left)**
  - **R (Right)**
- Compute the middle index using:

```python
M = (L + R) // 2
```

or the overflow-safe version:

```python
M = L + ((R - L) // 2)
```

- If the target is:
  - Equal → Return the index.
  - Smaller → Search the left half.
  - Larger → Search the right half.
- Stop when **L > R** if the target is not found.
- **Time Complexity:** **O(log n)**
- **Space Complexity (Iterative):** **O(1)**
- Binary Search is one of the most important algorithms for coding interviews and is frequently used as the foundation for solving problems involving sorted arrays.


````python
a = [-9,-7,-2,0,1,5,16,81,916,999]

#Traditional binary search

def binary_search(arr,target):

    L = 0
    N = len(arr)
    R = N -1

    while L <= R:
        M = (L + R) // 2
        if arr[M] == target:
           return M
        elif arr[M] < target:
          L = M + 1
        else:
            R = M - 1
    
    return False


binary_search(a,1000)
````

````python
# Condition based
B = [False, False, False, False, True]

def binary_search_condition(arr):
  N = len(arr)
  L = 0
  R = N - 1

  while L < R:
    M = (L + R) // 2

    if arr[M]:
      R = M
    else:
      L = M + 1

  return L

binary_search_condition(B)
````

````python
# range based 

def guess_number(target,L,R):
    while L <= R:
        M = (L+R)//2
        if M == target:
            return M
        elif M < target:
            L = M + 1
        else:
            R = M - 1
    return False


result = guess_number(5,0,2000)
print(f'Guessed num is {result}')
````
