```python
#Array

A = [1,2,3]
print(A)
```

```python
# Insert at the End - O(1)
A.append(7)
print(A)
```

```python
#Pop from the end - O(1)
A.pop() 
print(A)
```

```python
#Insert not at the end - O(n)
A.insert(0, 4) # insert at index
print(A)
```

```python
# Modift an element in a list - O(1)
A[0] = 8
print(A)
```

```python
# Accessing an element in a list - O(1)
print(f'Element in A = {A[3]}')
```

```python
if 8 in A: # 0(n)
    print('Yes')
```

```python
#Strings:

s = 'Hello World'
print(s)
```

```python
#Append to the end of a string - O(1)
s += "!"
print(s)
```

```python
# Accessing an element in a string - O(1)
print(f'Element in s = {s[6]}')
```

```python
#Check if an element is in string - O(n)
if '!' in s:
    print('Yes')
```

```python
# Check the length of the String - O(1)
print(len(s))
```
