# 📚 Sorting Algorithms

Sorting is the process of arranging data in a specific order (usually **ascending** or **descending**).

Example:

Before Sorting:

```
[8, 3, 5, 1, 9]
```

After Sorting:

```
[1, 3, 5, 8, 9]
```

---

# 🫧 Bubble Sort

## 📌 What is Bubble Sort?

Bubble Sort repeatedly compares **adjacent elements** and swaps them if they are in the wrong order.

With every pass, the **largest element "bubbles" to the end** of the array.

---

## 🔄 Bubble Sort Process

Example:

```
[5, 2, 4, 1]
```

### Pass 1

Compare:

```
5 2

↓

Swap
```

```
[2, 5, 4, 1]
```

Compare:

```
5 4

↓

Swap
```

```
[2, 4, 5, 1]
```

Compare:

```
5 1

↓

Swap
```

```
[2, 4, 1, 5]
```

Largest element (**5**) is now in its correct position.

Repeat until no swaps occur.

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Best Case | **O(n)** |
| Average Case | **O(n²)** |
| Worst Case | **O(n²)** |
| Space | **O(1)** |

---

## ✅ Characteristics

- In-place sorting
- Stable sorting algorithm
- Very simple to implement
- Inefficient for large datasets

---

# 🃏 Insertion Sort

## 📌 What is Insertion Sort?

Insertion Sort divides the array into:

- **Sorted Region**
- **Unsorted Region**

Each new element is picked from the unsorted region and inserted into its correct position in the sorted region.

---

## 🔄 Process

Example:

```
[5, 2, 4, 1]
```

Initially:

```
Sorted

[5]

Unsorted

[2,4,1]
```

Insert 2

```
[2,5]
```

Insert 4

```
[2,4,5]
```

Insert 1

```
[1,2,4,5]
```

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Best Case | **O(n)** |
| Average Case | **O(n²)** |
| Worst Case | **O(n²)** |
| Space | **O(1)** |

---

## ✅ Characteristics

- In-place
- Stable
- Performs well on nearly sorted arrays
- Frequently used for small datasets

---

# 🎯 Selection Sort

## 📌 What is Selection Sort?

Selection Sort repeatedly finds the **minimum element** from the unsorted portion and places it at the beginning.

---

## 🔄 Process

Example:

```
[5,3,4,1]
```

Find minimum:

```
1
```

Swap:

```
[1,3,4,5]
```

Move to next position.

Find minimum from remaining elements.

Repeat until sorted.

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Best Case | **O(n²)** |
| Average Case | **O(n²)** |
| Worst Case | **O(n²)** |
| Space | **O(1)** |

---

## ✅ Characteristics

- In-place
- Not stable (standard implementation)
- Performs the same number of comparisons regardless of input order

---

# 🌳 Merge Sort (Divide & Conquer)

## 📌 What is Merge Sort?

Merge Sort is a **Divide and Conquer** algorithm.

It repeatedly divides the array into smaller halves until each subarray contains only **one element**, then merges them back together in sorted order.

---

## 🔄 Divide Phase

Example:

```
[8,3,5,1]
```

↓

```
[8,3]

[5,1]
```

↓

```
[8]

[3]

[5]

[1]
```

Single-element arrays are already sorted.

---

## 🔀 Merge Phase

Merge:

```
[8]

[3]
```

↓

```
[3,8]
```

Merge:

```
[5]

[1]
```

↓

```
[1,5]
```

Final Merge:

```
[3,8]

[1,5]
```

↓

```
[1,3,5,8]
```

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Best Case | **O(n log n)** |
| Average Case | **O(n log n)** |
| Worst Case | **O(n log n)** |
| Space | **O(n)** (Typical Python implementation) |

---

## ✅ Characteristics

- Stable
- Very efficient
- Uses Divide & Conquer
- Requires additional memory in common implementations

---

# ⚡ Quick Sort

## 📌 What is Quick Sort?

Quick Sort selects a **Pivot** element and partitions the array into:

- Elements smaller than the pivot
- Pivot
- Elements larger than the pivot

The same process is then applied recursively to the left and right partitions.

---

## 🔄 Example

Array:

```
[8,3,5,1]
```

Choose Pivot:

```
1
```

Partition:

```
Left

[]

Pivot

[1]

Right

[8,3,5]
```

Recursively sort the right partition.

Eventually:

```
[1,3,5,8]
```

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Best Case | **O(n log n)** |
| Average Case | **O(n log n)** |
| Worst Case | **O(n²)** |
| Space | **O(log n)** (Recursion stack) |

---

## ⚠️ Worst Case

Occurs when poor pivots are repeatedly chosen.

Example:

```
[1,2,3,4,5]
```

Choosing:

```
1
```

every time creates extremely unbalanced partitions.

---

## ✅ Characteristics

- Very fast in practice
- Divide & Conquer algorithm
- Usually faster than Merge Sort due to cache efficiency
- Performance depends on pivot selection

---

# 🔢 Counting Sort

## 📌 What is Counting Sort?

Counting Sort is **not a comparison-based algorithm**.

Instead of comparing values, it counts how many times each value appears.

---

## 🔄 Process

Example:

```
[2,1,3,2,1]
```

Count Array:

```
Value

1 → 2

2 → 2

3 → 1
```

Rebuild:

```
[1,1,2,2,3]
```

---

## ⏱️ Complexity

| Operation | Complexity |
|-----------|:----------:|
| Time | **O(n + K)** |
| Space | **O(K)** |

Where:

- **n** = Number of elements
- **K** = Maximum possible value (range of input)

---

## ⚠️ Limitation

Counting Sort is efficient only when the range of values (**K**) is relatively small.

Example:

Good:

```
Values

0 - 100
```

Poor:

```
Values

0 - 1,000,000,000
```

The count array becomes extremely large, making the algorithm memory-intensive.

---

# 📊 Comparison of Sorting Algorithms

| Algorithm | Best | Average | Worst | Space | Stable | In-place |
|-----------|:----:|:-------:|:-----:|:-----:|:------:|:--------:|
| Bubble Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ | ✅ |
| Insertion Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ | ✅ |
| Selection Sort | O(n²) | O(n²) | O(n²) | O(1) | ❌ | ✅ |
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | O(n) | ✅ | ❌ |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | O(log n) | ❌ | ✅ |
| Counting Sort | O(n + K) | O(n + K) | O(n + K) | O(K) | ✅ | ❌ |

---

# 🎯 When to Use Each Algorithm

| Algorithm | Best Use Case |
|-----------|---------------|
| Bubble Sort | Learning algorithms; very small datasets |
| Insertion Sort | Small or nearly sorted arrays |
| Selection Sort | Simple implementation with minimal memory usage |
| Merge Sort | Large datasets requiring guaranteed **O(n log n)** performance and stable sorting |
| Quick Sort | General-purpose sorting with excellent average-case performance |
| Counting Sort | Sorting integers when the value range is small |

---

# 📝 Key Takeaways

### 🫧 Bubble Sort

- Compares **adjacent elements** and swaps them if needed.
- Largest element "bubbles" to the end after each pass.
- Time Complexity: **O(n²)**
- Space Complexity: **O(1)**

---

### 🃏 Insertion Sort

- Builds a **sorted region** one element at a time.
- Inserts each new element into its correct position.
- Time Complexity: **O(n²)**
- Performs well on nearly sorted arrays.

---

### 🎯 Selection Sort

- Repeatedly selects the **minimum element** from the unsorted portion.
- Swaps it into its correct position.
- Time Complexity: **O(n²)**

---

### 🌳 Merge Sort

- Uses the **Divide & Conquer** approach.
- Splits the array recursively and merges sorted halves.
- Time Complexity: **O(n log n)**
- Typical Python implementations use **O(n)** extra space.

---

### ⚡ Quick Sort

- Selects a **Pivot** and partitions the array.
- Average Time Complexity: **O(n log n)**
- Worst Case: **O(n²)** due to poor pivot selection.
- Very fast in practice.

---

### 🔢 Counting Sort

- Counts the occurrences of each value instead of comparing elements.
- Time Complexity: **O(n + K)**
- Best suited for **small ranges of integer values**.
- Can become memory-intensive when the value range (**K**) is very large.

````python
# Bubble Sort
# Time: O(n^2)
# Space: O(1)

A = [-5, 3, 2, 1, -3, -3, 7, 2, 2]

def bubble_sort(arr):
    n = len(arr)
    flag = True
    while flag:
        flag = False
        for i in range(1, n):
            if arr[i-1] > arr[i]:
                arr[i], arr[i-1] = arr[i-1], arr[i]
                flag = True


bubble_sort(A)
A
````

````python
# Insertion Sort
# Time: O(n^2)
# Space: O(1)

B = [-5, 3, 2, 1, -3, -3, 7, 2, 2]

def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n):
        for j in range(i, 0, -1):
            if arr[j-1] > arr[j]:
                arr[j], arr[j-1] = arr[j-1], arr[j]
            else:
                break
        
        
insertion_sort(B)
B
````

````python
# Selection Sort
# Time: O(n^2)
# Space: O(1)

C = [-3, 3, 2, 1, -5, -3, 7, 2, 2]

def selection_sort(arr):
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i],arr[min_index] = arr[min_index],arr[i]


selection_sort(C)
C
````

````python
# Merge Sort
# Time: O(n log n)
# Space: O(n) - Note: can be Log n, but this is harder to write


D = [-5, 3, 2, 1, -3, -3, 7, 2, 2]

def merge_sort(arr):
    n = len(arr)

    if n == 1:
        return arr

    middle = n // 2
    left = arr[:middle]
    right = arr[middle:]

    left = merge_sort(left)
    right = merge_sort(right)
    l,r =  0, 0
    left_len,right_len = len(left),len(right)

    sorted_arr = [0] * n
    i = 0

    while l < left_len and r < right_len:
        if left[l] <= right[r]:
            sorted_arr[i] = left[l]
            l += 1
        else:
            sorted_arr[i] = right[r]
            r += 1
        i += 1

    while l < left_len:
        sorted_arr[i] = left[l]
        l += 1
        i += 1

    while r < right_len:
        sorted_arr[i] = right[r]
        r += 1
        i += 1

    return sorted_arr


merge_sort(D)
````

````python
# Quick Sort
# Time: O(n log n) (Average case, technically Worst case is O(n^2))
# Space: O(n)

E = [-5, 3, 2, 1, -3, -3, 7, 2, 2]

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[-1] # last element as the pivot

    left = [x for x in arr[:-1] if x <= pivot]
    right = [x for x in arr[:-1] if x > pivot]

    left = quick_sort(left)
    right = quick_sort(right)

    return left + [pivot] + right


quick_sort(E)
E
````

````python
# Counting Sort
# Time: O(n + k) where k is the range of data

# Note - This can be written with negative arrays, but we'll stick to positive arrays,
# so k is the max of the array

# Space: O(k)

F = [5, 3, 2, 1, 3, 3, 7, 2, 2]

def counting_sort(arr):
    n = len(arr)
    maxx = max(arr)
    counts = [0] * (maxx +1)

    for x in arr:
        counts[x] += 1

    i = 0
    
    for c in range(maxx +1):
        while counts[c] > 0:
            arr[i] = c
            i += 1
            counts[c] -= 1


counting_sort(F)
F
````
